```python
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

# numpy pandas matplotlib seaborn
# scikit-learn
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.info()
```

```python
sns.jointplot(x="tip", y="total_bill", data=df, kind="reg");
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df[["total_bill"]]
y = df[["tip"]]
X.shape, y.shape
```

```python
lr = LinearRegression() # y = b0 + b1*x
```

```python
model = lr.fit(X, y)
```

```python
model.intercept_ # b0
```

```python
model.coef_ # b1
```

```python
# r kare skoru (model doğruluğunu % olarak verir)
r_kare = model.score(X, y)
print("Doğruluk: %{:.2f}".format(r_kare*100))
```

```python
x = 15 # x = tahmin için girilen değer
tahmin = model.intercept_ + model.coef_*x  # y = b0 + b1*x
tahmin
```

```python
model.predict([[x]])
```

```python
bill = [[12], [30], [45]]
model.predict(bill)
```
